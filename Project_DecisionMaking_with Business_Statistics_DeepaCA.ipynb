**Problem 2**

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

**2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?**

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
mu = 5  # Mean breaking strength (kg/sq cm)
sigma = 1.5  # Standard deviation (kg/sq cm)
x = 3.17  # Breaking strength value

# Calculate the Z-score
z = (x - mu) / sigma

# Calculate the proportion (cumulative probability) for less than x
proportion = stats.norm.cdf(x, loc=mu, scale=sigma)

print(f"Mean (mu): {mu} kg/sq cm")
print(f"Standard Deviation (sigma): {sigma} kg/sq cm")
print(f"Value (x): {x} kg/sq cm")
print(f"Z-score: {z:.4f}")
print(f"2.1 Proportion of gunny bags with breaking strength less than {x} kg/sq cm: {proportion:.4f}")

# --- Visual Representation ---

# Generate a range of x values for the normal distribution curve
x_values = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)

# Calculate the probability density function (PDF) for these x values
pdf_values = stats.norm.pdf(x_values, loc=mu, scale=sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_values, pdf_values, label='Normal Distribution PDF')

# Shade the area less than x
x_fill = np.linspace(mu - 4 * sigma, x, 500)
y_fill = stats.norm.pdf(x_fill, loc=mu, scale=sigma)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label=f'Proportion < {x}')

# Add labels and title
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Gunny Bag Breaking Strength')
plt.axvline(x=mu, color='red', linestyle='--', label=f'Mean (mu) = {mu}')
plt.axvline(x=x, color='green', linestyle=':', label=f'x = {x}')
plt.legend()
plt.grid(True)
plt.show()
```

Answer: 2.1 Proportion of gunny bags with breaking strength less than 3.17 kg/sq cm: **0.1112**

**2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?**

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Given parameters (from previous problem)
mu = 5  # Mean breaking strength (kg/sq cm)
sigma = 1.5  # Standard deviation (kg/sq cm)
x = 3.6  # Breaking strength value

# Calculate the Z-score
z = (x - mu) / sigma

# Calculate the proportion (cumulative probability) for less than x
proportion_less_than_x = stats.norm.cdf(x, loc=mu, scale=sigma)

# Calculate the proportion for at least x (1 - P(X < x))
proportion_at_least = 1 - proportion_less_than_x

print(f"Mean (mu): {mu} kg/sq cm")
print(f"Standard Deviation (sigma): {sigma} kg/sq cm")
print(f"Value (x): {x} kg/sq cm")
print(f"Z-score: {z:.4f}")
print(f"2.2 Proportion of gunny bags with breaking strength at least {x} kg/sq cm: {proportion_at_least:.4f}")

# --- Visual Representation ---

# Generate a range of x values for the normal distribution curve
x_values = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 500)

# Calculate the probability density function (PDF) for these x values
pdf_values = stats.norm.pdf(x_values, loc=mu, scale=sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_values, pdf_values, label='Normal Distribution PDF')

# Shade the area at least x
x_fill = np.linspace(x, mu + 4 * sigma, 500)
y_fill = stats.norm.pdf(x_fill, loc=mu, scale=sigma)
plt.fill_between(x_fill, y_fill, color='lightcoral', alpha=0.5, label=f'Proportion >= {x}')

# Add labels and title
plt.xlabel('Breaking Strength (kg/sq cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Gunny Bag Breaking Strength')
plt.axvline(x=mu, color='red', linestyle='--', label=f'Mean (mu) = {mu}')
plt.axvline(x=x, color='green', linestyle=':', label=f'x = {x}')
plt.legend()
plt.grid(True)
plt.show()
```

**2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?**

**3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?**

To assess this, perform a one-sample t-test on the 'Unpolished' stones' hardness against the required minimum of 150.

Hypothesis:

**Null Hypothesis (H0):** The true mean hardness of unpolished stones is greater than or equal to 150.

**Alternative Hypothesis (H1):** The true mean hardness of unpolished stones is less than 150.

H0: Mean hardness of unpolished stones >= 150

H1: Mean hardness of unpolished stones < 150

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Hardness threshold
hardness_threshold = 150

# Extract data for unpolished stones
unpolished_hardness = df['Unpolished']

# --- Step 1: Formulate Hypotheses ---
# Null Hypothesis (H0): The true mean hardness of unpolished stones is greater than or equal to 150.
# Alternative Hypothesis (H1): The true mean hardness of unpolished stones is less than 150.
# This is a one-tailed test.

# Significance level
significance_level = 0.05

print(f"Hardness Threshold: {hardness_threshold}")
print(f"Significance Level: {significance_level}\n")
print("Hypotheses:")
print("H0: Mean hardness of unpolished stones >= 150")
print("H1: Mean hardness of unpolished stones < 150\n")

# --- Step 2: Perform One-Sample T-test ---
# We perform a one-sample t-test since we are comparing the mean of a single sample
# to a known constant (the threshold of 150).
# Since stats.ttest_1samp performs a two-tailed test, we need to adjust the p-value for a one-tailed test.

t_statistic, p_value = stats.ttest_1samp(unpolished_hardness, hardness_threshold, alternative='less')

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}\n")

# --- Step 3: Interpret Results ---
if p_value < significance_level:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({significance_level}), we reject the null hypothesis.")
    print("Conclusion: Zingaro is justified in thinking that the unpolished stones may not be suitable for printing, as their mean hardness is significantly less than 150.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({significance_level}), we fail to reject the null hypothesis.")
    print("Conclusion: Zingaro is not justified in thinking that the unpolished stones may not be suitable for printing, as there is no significant evidence that their mean hardness is less than 150.")

# --- Step 4: Visual Representation ---
plt.figure(figsize=(10, 6))
sns.histplot(unpolished_hardness, kde=True, color='skyblue', bins=15)
plt.axvline(x=hardness_threshold, color='red', linestyle='--', label=f'Hardness Threshold = {hardness_threshold}')
plt.title('Distribution of Unpolished Stone Hardness with Threshold')
plt.xlabel('Brinell Hardness Index')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Answer:

T-statistic: -4.1646
P-value: 0.0000

Since the p-value (0.0000) is less than the significance level (0.05), we reject the null hypothesis.

Conclusion: Zingaro is justified in thinking that the unpolished stones may not be suitable for printing, as their mean hardness is significantly less than 150.

**3.2 Is the mean hardness of the polished and unpolished stones the same?**

To assess this, perform an independent two-sample t-test to compare the means of the two groups (unpolished and polished stones).

Hypothesis:

Null Hypothesis (H0): The mean hardness of unpolished stones is equal to the mean hardness of polished stones.

Alternative Hypothesis (H1): The mean hardness of unpolished stones is not equal to the mean hardness of polished stones.

This is a two-tailed test.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Extract data for polished and unpolished stones
unpolished_hardness = df['Unpolished']
polished_hardness = df['Treated and Polished']

# --- Step 1: Formulate Hypotheses ---
# Null Hypothesis (H0): The mean hardness of unpolished stones is equal to the mean hardness of polished stones.
# Alternative Hypothesis (H1): The mean hardness of unpolished stones is not equal to the mean hardness of polished stones.
# This is a two-tailed test.

# Significance level (from problem description)
significance_level = 0.05

print(f"Significance Level: {significance_level}\n")
print("Hypotheses:")
print("H0: Mean hardness of unpolished stones = Mean hardness of polished stones")
print("H1: Mean hardness of unpolished stones != Mean hardness of polished stones\n")

# --- Step 2: Perform Independent Two-Sample T-test ---
# We perform an independent t-test as the two groups (unpolished and polished) are distinct.
# We assume equal variances for now; a Levene's test could confirm this, but ttest_ind often handles it well.

t_statistic, p_value = stats.ttest_ind(unpolished_hardness, polished_hardness, equal_var=True)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}\n")

# --- Step 3: Interpret Results ---
if p_value < significance_level:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({significance_level}), we reject the null hypothesis.")
    print("Conclusion: The mean hardness of unpolished stones is significantly different from the mean hardness of polished stones.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({significance_level}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no significant evidence to suggest that the mean hardness of unpolished stones is different from the mean hardness of polished stones.")

# --- Step 4: Visual Representation ---
plt.figure(figsize=(10, 6))
sns.histplot(unpolished_hardness, kde=True, color='skyblue', label='Unpolished', bins=15)
sns.histplot(polished_hardness, kde=True, color='lightcoral', label='Polished', bins=15, alpha=0.7)
plt.axvline(x=unpolished_hardness.mean(), color='blue', linestyle='--', label=f'Mean Unpolished = {unpolished_hardness.mean():.2f}')
plt.axvline(x=polished_hardness.mean(), color='red', linestyle='--', label=f'Mean Polished = {polished_hardness.mean():.2f}')
plt.title('Distribution of Unpolished vs. Polished Stone Hardness')
plt.xlabel('Brinell Hardness Index')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Answer:

**4.1 How does the hardness of implants vary depending on dentists?**

Perform a One-Way ANOVA test. This will help us determine if there's a statistically significant difference in mean hardness across the different dentists.

Hypothesis:

 **Null Hypothesis (H0):** The mean hardness of implants is the same across all dentists.

 **Alternative Hypothesis (H1):** At least one dentist's mean hardness is different from the others.


**Assumptions of test:**

The assumptions of the One-Way ANOVA test, which was used in section 4.1 to determine how the hardness of implants varies depending on dentists, are as follows:

**Independence of Observations:** The observations within each group (dentist) must be independent of each other, and the observations between groups must also be independent.

**Normality:** The dependent variable (implant hardness/Response) should be approximately normally distributed for each group (each dentist). ANOVA is relatively robust to minor deviations from normality, especially with larger sample sizes.

**Homogeneity of Variances:** The variance of the dependent variable (implant hardness) should be approximately equal across all groups (all dentists). This is often tested using Levene's test.

**For testing of normality, Shapiro-Wilk’s test is applied**

Let us do Shapiro Wilk's test, to verify assumptions of ANOVA test

```python
# Library for statistical analysis
import scipy.stats as stats
for id in df1['Dentist'].unique():
    # Extract response data
    response_data = df1[df1['Dentist'] == id]['Response']

    # Perform Shapiro-Wilk test
    stat, p = stats.shapiro(response_data)

    print(f"Dentist {id}: Statistics={stat:.3f}, p={p:.3f}")
    if p < 0.05:
        print(f"  -> Sample does not look Gaussian (reject H0 for normality)")
    else:
        print(f"  -> Sample looks Gaussian (fail to reject H0 for normality)")
```

**Results of Shapiro Wilk's test:**

Dentist 4's data significantly deviates from normality. The p-value is less than 0.05, leading us to reject the null hypothesis of normality for this group.

All other Dentist's show's normal distribution.The p-value is greater than 0.05, leading us to accept the null hypothesis of normality for this group.

However, ANOVA is generally robust to mild violations of normality.
Hence we will progress with ANOVA.



**For equality of variance, Levene's test is applied**

Let us do Levene's test, to verify assumptions of ANOVA test

**4.4 How does the hardness of implants vary depending on dentists and methods together?**

A two-way ANOVA will be performed using statsmodels to analyze the effect of 'Dentist', 'Method', and their interaction on 'Response' (hardness).

The p-values from the ANOVA for the interaction term (Dentist:Method) will be examined for each alloy to determine if a statistically significant interaction effect exists at a 5% significance level.

**Hypothesis:**

**Null Hypothesis ( H0 ):**

There is no interaction effect between Dentist and Method on the mean hardness of dental implants.

**Alternative Hypothesis ( H1 ):**
There is a statistically significant interaction effect between Dentist and Method on the mean hardness of dental implants.

**Assumptions of Test:**


1. Independence of Observations: All observations within and between groups must be independent.

2. Normality: The dependent variable (implant hardness/Response) should be approximately normally distributed for each combination of the independent variables (each Dentist-Method combination within each Alloy type). ANOVA is generally robust to moderate violations of normality, especially with larger sample sizes.

3. Homogeneity of Variances (Homoscedasticity): The variance of the dependent variable should be approximately equal across all groups (i.e., for all combinations of Dentist and Method within each Alloy type). Levene's test is commonly used to check this assumption.

4. Categorical Independent Variables: The independent variables ('Dentist' and 'Method') must be categorical (nominal or ordinal).

5. Continuous Dependent Variable: The dependent variable ('Response' / hardness) must be continuous and measured on an interval or ratio scale.


To check the interaction effect between 'Dentist' and 'Method' on the hardness of dental implants for each type of alloy, we will prepare Data and Perform ANOVA for Each Alloy.

For each unique 'Alloy' type, the data will be filtered. Then, a two-way ANOVA will be performed using `statsmodels` to analyze the effect of 'Dentist', 'Method', and their interaction on 'Response' (hardness).

The p-values from the ANOVA for the interaction term (Dentist:Method) will be examined for each alloy to determine if a statistically significant interaction effect exists at a 5% significance level.

**Shapiro-Wilk’s test**

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
for alloy_id in df1['Alloy'].unique():
    print(f"\n--- Analyzing for Alloy {alloy_id} ---")
    alloy_df = df1[df1['Alloy'] == alloy_id]

    # Shapiro-Wilk Test for Normality (on residuals for more rigorous check, or on groups for simplicity)
    # For simplicity, we'll check normality for each Dentist-Method group within this Alloy
    print(f"Shapiro-Wilk Test for Normality for Alloy {alloy_id} (by Dentist-Method Group):")
    for dentist_id in alloy_df['Dentist'].unique():
        for method_id in alloy_df['Method'].unique():
            group_data = alloy_df[(alloy_df['Dentist'] == dentist_id) & (alloy_df['Method'] == method_id)]['Response']
            if len(group_data) > 3: # Shapiro-Wilk requires at least 3 data points
                stat, p = stats.shapiro(group_data)
                print(f"  Dentist {dentist_id}, Method {method_id}: Statistics={stat:.3f}, p={p:.3f}")
                if p <= 0.05:
                    print("    -> Group does not look Gaussian (Reject H0 for normality)")
                # else:
                #     print("    -> Group looks Gaussian (Fail to reject H0 for normality)")
            else:
                print(f"  Not enough data points for Dentist {dentist_id}, Method {method_id}. Skipping Shapiro-Wilk test")

```

**Levene's Test**

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

for alloy_id in df1['Alloy'].unique():
    print(f"\n--- Analyzing for Alloy {alloy_id} ---")
    alloy_df = df1[df1['Alloy'] == alloy_id]
# Create groups based on interaction of Dentist and Method
    groups_for_levene = []
    for dentist_id in alloy_df['Dentist'].unique():
        for method_id in alloy_df['Method'].unique():
            group_data = alloy_df[(alloy_df['Dentist'] == dentist_id) & (alloy_df['Method'] == method_id)]['Response']
            if not group_data.empty:
                groups_for_levene.append(group_data)

    if len(groups_for_levene) > 1:
        stat, p = stats.levene(*groups_for_levene)
        print(f"  Statistics={stat:.3f}, p={p:.3f}")
        if p <= 0.05:
            print("    -> Variances are not equal (Reject H0 for equal variances)")
        else:
            print("    -> Variances are equal (Fail to reject H0 for equal variances)")
    else:
        print("  Not enough groups to perform Levene's test for this alloy.")
```

**ANOVA TEST**

ANOVA is often robust to mild deviations from normality, especially when sample sizes are balanced and variances are equal. So let us progress, even though assumptions for normality are not fully satisfied.

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

# Identify unique alloy types
unique_alloys = df1['Alloy'].unique()

print("--- Analyzing Interaction Effect for Each Alloy Type ---")

# Iterate through each unique 'Alloy' type
for alloy_type in unique_alloys:
    print(f"\nAlloy Type: {alloy_type}")

    # Filter the DataFrame for the current 'Alloy' type
    df_alloy = df1[df1['Alloy'] == alloy_type]

    # Construct the OLS model formula for two-way ANOVA with interaction
    # C() treats the variables as categorical
    model_formula = 'Response ~ C(Dentist) + C(Method) + C(Dentist):C(Method)'
    print(f"ANOVA Model Formula: {model_formula}")

    # Create and fit the OLS model
    model = smf.ols(model_formula, data=df_alloy).fit()

    # Perform ANOVA on the fitted model
    anova_table = sm.stats.anova_lm(model, typ=2) # typ=2 for Type II sum of squares

    print("ANOVA Results:")
    print(anova_table)

    p_value_interaction = anova_table.loc['C(Dentist):C(Method)', 'PR(>F)']

    print(f"P-value for interaction (Dentist:Method) for Alloy {alloy_id}: {p_value_interaction:.4f}")

    alpha = 0.05

    if p_value_interaction < alpha:
        print(f"Conclusion for Alloy {alloy_id}: Reject the null hypothesis. There is a statistically significant interaction effect between Dentist and Method on implant hardness.")
    else:
        print(f"Conclusion for Alloy {alloy_id}: Fail to reject the null hypothesis. There is NO statistically significant interaction effect between Dentist and Method on implant hardness.")



print("\n--- Visual Representation of Interaction Effect ---")
# For interaction effects, it's often best to visualize with interaction plots
# Plotting interaction effect for each alloy type
for alloy_type in unique_alloys:
    plt.figure(figsize=(10, 6))
    sns.pointplot(x='Dentist', y='Response', hue='Method', data=df1[df1['Alloy'] == alloy_type],palette='plasma', dodge=True)
    plt.title(f'Interaction Plot of Dentist and Method on Hardness for Alloy {alloy_type}')
    plt.xlabel('Dentist')
    plt.ylabel('Hardness (Response)')
    plt.legend(title='Method')
    plt.grid(True)
    plt.show()
```
